```python
def bootstrap_resample(X, n = None):
    """ Bootstrap resample an array_like
    Parameters
    ----------
    X : array_like
      data to resample
    n : int, optional
      length of resampled array, equal to len(X) if n==None
    Results
    -------
    returns X_resamples
    Source: http://nbviewer.ipython.org/gist/aflaxman/6871948
    """
    if n == None:
        n = len(X)
        
    resample_i = np.floor(np.random.rand(n)*len(X)).astype(int)
    X_resample = X[resample_i]
    return X_resample
```

```python
train_performance_by_minute = pd.DataFrame(index = range(60), columns = ['rf_full', 'rf_limited', 'gbm'])
test_performance_by_minute = pd.DataFrame(index = range(60), columns = ['rf_full', 'rf_limited', 'gbm'])
```

```python
seconds = 3600
index = 0

while seconds > 0:
    train_min_features = df[(df['sec_adj'] > seconds - 100) & (df['sec_adj'] <= seconds)][['sec_adj', 'sec_adj_sqrt', 'dwn', 'ytg', 'yfog', 'score_diff', 'sprv', 'ou']]
    test_min_features = test_df[(test_df['sec_adj'] > seconds - 100) & (test_df['sec_adj'] <= seconds)][['sec_adj', 'sec_adj_sqrt', 'down', 'togo', 'yfog', 'scorediff', 'sprv', 'ou']]
    
    target_min = df[(df['sec_adj'] > seconds - 100) & (df['sec_adj'] <= seconds)]['win']
    test_target_min = test_df[(test_df['sec_adj'] > seconds - 100) & (test_df['sec_adj'] <= seconds)]['teamwin']
    
    train_performance_by_minute.ix[index, 'rf_full'] = model_rf.score(train_min_features, target_min)
    test_performance_by_minute.ix[index, 'rf_full'] = model_rf.score(test_min_features, test_target_min)
    
    train_performance_by_minute.ix[index, 'rf_limited'] = model_limited.score(train_min_features[['sec_adj', 'dwn', 'ytg', 'yfog', 'score_diff']], target_min)
    test_performance_by_minute.ix[index, 'rf_limited'] = model_limited.score(test_min_features[['sec_adj', 'down', 'togo', 'yfog', 'scorediff']], test_target_min)
    
    train_performance_by_minute.ix[index, 'gbm'] = model_gbm.score(train_min_features, target_min)
    test_performance_by_minute.ix[index, 'gbm'] = model_gbm.score(test_min_features, test_target_min)
    
    seconds -= 60
    index += 1
```

```python
performance = pd.concat([train_performance_by_minute, test_performance_by_minute], axis = 1)
performance.columns = ['rf_full_train', 'rf_limited_train', 'gbm_train', 'rf_full_test', 'rf_limited_test', 'gbm_test']
```
